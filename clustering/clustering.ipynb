## Clustering

```python
from chunktext import chunk_txt
```

```python
chunks = chunk_txt("himling.txt")
```

```python
from embeddings import embed_txt
```

```python
model="text-embedding-ada-002"
```

```python
def embed_chunk(text, model):
    response = client.embeddings.create(
        model=model,
        input=text
    )
    return response.data[0].embedding
```

```python
chunks_embeddings = np.array([np.array(embed_chunk(chunk, model)) for chunk in chunks])
```

```python
chunks_embeddings.shape
```

```python
from sklearn.cluster import KMeans

n_clusters = 4

kmeans = KMeans(n_clusters=n_clusters, init="k-means++", random_state=42)
kmeans_test = kmeans.fit(chunks_embeddings)
```

```python
kmeans_test.labels_
```

```python
cluster_elements = [chunks[i] for i in range(len(chunks)) if kmeans_test.labels_[i] == 2]

# Sjekker cluster med label 0:
for element in cluster_elements:
    print(" ".join(element), "\n\n")
```
